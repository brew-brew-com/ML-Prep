```python
X = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
y = pd.DataFrame(breast_cancer.target, columns=["class"])

print("X Shape: [%i, %i]" %X.shape)
print("y Shape: [%i, %i]" %y.shape)

display(X.join(y).head(10))
```
